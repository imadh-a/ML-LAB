```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/ML LAB FINAL/Candidate Elimination/enjoysport_dataset_1BM20CS059.csv')
print(data, "\n")

concepts = np.array(data.iloc[:, :-1])
print("The attributes are: ", concepts)
target = np.array(data.iloc[:, -1])
print("\nThe target is: ", target)

specific = concepts[0].copy()
general = [['?' for _ in range(len(specific))] for _ in range(len(specific))]

for i in range(len(concepts)):
    if target[i] == "yes":
        for j in range(len(specific)):
            if concepts[i][j] != specific[j]:
                specific[j] = '?'
                general[j][j] = '?'
    else:
        for j in range(len(specific)):
            if concepts[i][j] != specific[j]:
                general[j][j] = specific[j]

    print("\nStep " + str(i + 1) + " of Candidate Elimination Algorithm")
    print("Specific Hypothesis:", specific)
    print("General Hypothesis:", general)

gh = []
for hypothesis in general:
    if '?' not in hypothesis:
        gh.append(hypothesis)

print("\nFinal Specific hypothesis:\n", specific)
print("\nFinal General hypothesis:\n", gh)

def predict(instance, specific_hypothesis, general_hypotheses):
    for i in range(len(specific_hypothesis)):
        if specific_hypothesis[i] != '?' and specific_hypothesis[i] != instance[i]:
            return "no"

    for gh in general_hypotheses:
        match = True
        for i in range(len(gh)):
            if gh[i] != '?' and gh[i] != instance[i]:
                match = False
                break
        if match:
            return "yes"

    return "unknown"

predicted_targets = []
for instance in concepts:
    prediction = predict(instance, specific, gh)
    predicted_targets.append(prediction)

def calculate_accuracy(actual, predicted):
    correct = sum(1 for a, p in zip(actual, predicted) if a == p)
    total = len(actual)
    accuracy = correct / total * 100
    return accuracy

# Convert target values to numeric (0 and 1)
target_numeric = np.where(target == "yes", 1, 0)
predicted_numeric = np.where(np.array(predicted_targets) == "yes", 1, 0)

# Calculate accuracy
accuracy = calculate_accuracy(target_numeric, predicted_numeric)
print("\nAccuracy:", accuracy)
```
